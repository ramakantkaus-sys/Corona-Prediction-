```python
from sklearn.linear_model import LinearRegression
```

```python
x = data[['Age','Fever','Cough','Shortness of Breath','Contact with Confirmed']]
x
```

```python
y = data['COVID Positive']
y
```

```python
from sklearn.model_selection import train_test_split

# Splitting data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)  # Predicted salaries
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# R-squared value (explains how well the model fits)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f'R-squared: {r2:.2f}')
print(f'Mean Squared Error: {mse:.2f}')
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
```

```python
ridge = Ridge(alpha=1.0)  # Alpha is the regularization strength
ridge.fit(x_train, y_train)

# Predicting
y_pred_ridge = ridge.predict(x_test)

# Evaluate Ridge Regression
ridge_mse = mean_squared_error(y_test, y_pred_ridge)
ridge_r2 = r2_score(y_test, y_pred_ridge)

print(f'Ridge Regression - MSE: {ridge_mse:.2f}, R²: {ridge_r2:.2f}')
```

```python
lasso = Lasso(alpha=0.1)  # Smaller alpha keeps more features
lasso.fit(x_train, y_train)

# Predicting
y_pred_lasso = lasso.predict(x_test)

# Evaluate Lasso Regression
lasso_mse = mean_squared_error(y_test, y_pred_lasso)
lasso_r2 = r2_score(y_test, y_pred_lasso)

print(f'Lasso Regression - MSE: {lasso_mse:.2f}, R²: {lasso_r2:.2f}')
```

```python
print("Ridge Coefficients:", ridge.coef_)
print("Lasso Coefficients:", lasso.coef_)
```
